```python
import hashlib
import requests
import json

API_UID = "TDQH1GD"
API_KEY = "FK1KE4T61D3A92A18KLBG3R2V11GKHA7"
BASE_URL = "https://bp.autoiso.pl/api/v3"

def decode_vin(vin: str, lang: str = "en") -> dict:
    # Generate MD5 checksum: MD5(apiuid + api_key + vin)
    checksum_str = API_UID + API_KEY + vin
    checksum = hashlib.md5(checksum_str.encode()).hexdigest()

    url = f"{BASE_URL}/getDecoderFree/apiuid:{API_UID}/checksum:{checksum}/vin:{vin}/lang:{lang}"
    
    print(f"\nRequesting: {url}\n")
    
    response = requests.get(url)
    response.raise_for_status()
    
    return response.json()

```
